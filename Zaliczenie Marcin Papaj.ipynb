```python
import os 
from openpyxl import load_workbook
import pandas as pd
import matplotlib.pyplot as plt
excel_files=[]
student_set=set()
sub_name=set()
directory=os.listdir()
#marks=pd.DataFrame()




def subject_names():
    find_excel_files(directory)
    global marks
    marks=pd.DataFrame()
    #print(excel_files)
    worksheets=[]
    for files in excel_files:
        
        wb=load_workbook(filename=files)
        worksheets=wb.sheetnames
        sub_name.update(worksheets)
        #creating list of students
        for sub in worksheets:
            for student in wb[sub]['A:A']:
                student_set.add(student.value)
    

    marks=pd.DataFrame(columns=sub_name,index=student_set)
                
def filling_marks():
    for files in excel_files:
        wb=load_workbook(filename=files)
        worksheets=wb.sheetnames
        for sheet in worksheets:
            
            subj=wb[sheet]
            for index,row in enumerate(subj.iter_rows(min_row=1,max_col=1)):
                name=subj.cell(row=index+1,column=1).value
                grade=subj.cell(row=index+1,column=2).value
                if name in marks.index:
                    marks.loc[[name],sheet]=grade
                
                



def find_excel_files(a):
    
    for file in a:
        if file.endswith(".xlsx"):
            excel_files.append(file)
    
    
def calculateAverageForSubject(subject):
    
    srednia=marks[subject].mean()
    return ("Srednia ocen dla przedmiotu {} wynosi {:.2f}".format(subject,srednia))
    
        
def calculateAverageForStudent(student):
    
    srednia=marks.loc[student].mean()
    return ("Srednia ocen dla studenta {} wynosi {:.2f}".format(student,srednia))


def calculateAverageForAllSubjects(plot=True):
    
    srednia_dla_studentow=[]
   
    for row in marks.index:
        
        
        srednia_dla_studentow.append(round(marks.loc[row].mean(),2))
    try:
        
        marks.insert(3,"Srednia dla studentów",srednia_dla_studentow)
    except:
        pass
    
    if plot:
        
        #plt.figure(figsize=(26,2))
        plt.figure(figsize=(30, 5))
        
        #ax = fig.add_axes([5,5,5,1])
        #plot=plt.bar(marks.index,marks["Srednia dla studentów"].sort_values(ascending=False))
        
        plt.bar(marks.index,marks["Srednia dla studentów"].sort_values(ascending=False))
        plt.title('Srednia ocen dla studentów')
        
        
        
    return ("Średnia dla wszystkich studentów wynosi {:.2f}".format(marks["Srednia dla studentów"].mean()))
        
def generateStudentRankingForSubject(subject):
    #marks.sort_values(by=[subject])
    
    print(marks[subject].sort_values(ascending=False))
    

def generateStudentRankingForAllSubjects():
    calculateAverageForAllSubjects(False)
    
    return (marks["Srednia dla studentów"].sort_values(ascending=False))
    
    
    
    

if __name__ == "__main__":
    #Wywołanie funkcji wypełniających dataframe, tak aby nie powtarzać kodu w każdej funkcji
    subject_names()
    filling_marks() 


generateStudentRankingForAllSubjects()

   
    
```
